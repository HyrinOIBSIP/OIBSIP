# ***Checking the correlation between different features in the dataset***

```python
#quantify and visualize the relationships between variables in a dataset, aiding in exploratory data analysis, feature selection, and model building.
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(11,6))
ax=sns.heatmap(df_2020.corr(),annot=True,linewidth=3)
#This line creates a heatmap of the correlation matrix for the DataFrame df_2020, adding annotations (correlation coefficients) and setting the line width between cells.
ax.tick_params(size=10,color='w',labelsize=10)
#adjusts the size, color, and label size of the tick marks (axis labels) on the heatmap plot to enhance readability.
plt.show()
```
